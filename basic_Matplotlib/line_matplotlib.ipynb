```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
years = [2008 + n for n in range(17)]
print(len(years))
```

```python
weight = np.random.randint(60, 70, 17)
print(len(weight))
```

```python
plt.plot(years, weight)
plt.show()
```

```python
plt.plot(years, weight, c="red", alpha=0.5)
plt.show()
```

```python
# add title and x,y levels 
plt.plot(years, weight, c="red", alpha=0.8)
plt.xlabel("years")
plt.ylabel("weights")
plt.title("Line Graph")

plt.show()
```

```python
# change the line width and line style 
# method 1
plt.plot(years, weight, c="red", alpha=0.8, lw=3, linestyle="--")
plt.xlabel("years")
plt.ylabel("weights")
plt.title("Line Graph")

plt.show()
```

```python
# change the line width and line style 
# method 2
plt.plot(years, weight, "r--", lw=3) # shorhand for line color and line style
plt.xlabel("years")
plt.ylabel("weights")
plt.title("Line Graph")

plt.show()
```
