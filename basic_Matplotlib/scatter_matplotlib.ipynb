```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# create a dataset
x_data = np.random.random(50) * 100
# print(x_data)

y_data = np.random.random(50) * 100
# print(y_data)
```

```python
# scatter plot
plot = plt.scatter(x_data, y_data)
```

```python
# change circle color
plt.scatter(x=x_data, y=y_data, c="red")
plt.show()
```

```python
# chnage circle to star 
plot = plt.scatter(x=x_data, y=y_data, c="green", marker="*")
```

```python
# added title
plot = plt.scatter(x_data, y_data, c="green", marker="*")
plt.title("Scatter Plot")
```

```python
# add x,y levels
plt.scatter(x_data, y_data, c="g", marker="*")
plt.xlabel(xlabel="X axis")
plt.ylabel("Y axis")

plt.show()
```

```python
# alpha,
plt.scatter(x_data, y_data, c="g", marker="*", alpha=0.5) 
plt.xlabel(xlabel="X axis")
plt.ylabel("Y axis")

plt.show()
```

```python
# size of star/circle etc
plt.scatter(x_data, y_data, c="g", marker="*", s=220)
plt.xlabel(xlabel="X axis")
plt.ylabel("Y axis")

plt.show()
```
